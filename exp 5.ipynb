```python
import numpy as np

# Example fuel efficiency data (mpg)
# Replace with your actual dataset
fuel_efficiency = np.array([22, 25, 28, 30, 35])

# 1️⃣ Calculate average fuel efficiency
average_efficiency = np.mean(fuel_efficiency)

# 2️⃣ Percentage improvement between two car models
# Example: compare model at index 1 and index 4
model_1 = fuel_efficiency[1]   # Old model
model_2 = fuel_efficiency[4]   # New model

# Percentage Improvement Formula
# ((New - Old) / Old) * 100

if model_1 != 0:
    percentage_improvement = ((model_2 - model_1) / model_1) * 100
else:
    percentage_improvement = 0

# Print results
print("Average Fuel Efficiency:", average_efficiency, "mpg")
print("Percentage Improvement:", percentage_improvement, "%")
```
