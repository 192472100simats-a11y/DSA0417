```python
# Import libraries
import numpy as np
import scipy.stats as stats

# Example blood pressure reduction data (mmHg)
# Drug group (25 patients)
drug = [12, 15, 10, 14, 13, 11, 16, 12, 14, 15,
        13, 12, 17, 16, 14, 13, 12, 15, 16, 14,
        13, 12, 14, 15, 16]

# Placebo group (25 patients)
placebo = [5, 6, 4, 5, 3, 6, 5, 4, 6, 5,
           4, 5, 3, 6, 4, 5, 4, 6, 5, 3,
           4, 5, 6, 4, 5]

# Function to calculate 95% confidence interval
def confidence_interval(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    n = len(data)
    ci = stats.t.interval(0.95, df=n-1, loc=mean, scale=std/np.sqrt(n))
    return mean, ci

# Calculate CI for Drug group
drug_mean, drug_ci = confidence_interval(drug)

# Calculate CI for Placebo group
placebo_mean, placebo_ci = confidence_interval(placebo)

# Print results
print("Drug Group Mean Reduction:", drug_mean)
print("95% Confidence Interval (Drug):", drug_ci)

print("\nPlacebo Group Mean Reduction:", placebo_mean)
print("95% Confidence Interval (Placebo):", placebo_ci)
```
