```python
prices = [100, 50, 30]
quantities = [2, 3, 1]
discount_rate = 10
tax_rate = 5

subtotal = sum([p*q for p, q in zip(prices, quantities)])
total_after_discount = subtotal - (discount_rate/100)*subtotal
total_cost = total_after_discount + (tax_rate/100)*total_after_discount

print(total_cost)
```
