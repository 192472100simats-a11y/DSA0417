```python
from collections import Counter

# Customer reviews
reviews = [
    "Product is good",
    "Good quality product",
    "This product is very good"
]

# Convert reviews into words
text = " ".join(reviews).lower()
words = text.split()

# Frequency distribution
freq = Counter(words)

# Display result
print("Word Frequency:")
for word, count in freq.items():
    print(word, ":", count)
```
